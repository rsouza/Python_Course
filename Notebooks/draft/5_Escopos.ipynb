```python
x = 'global'
def local1():
    x = 'local1'
    def local2():
        x = 'local2'
        print(x)
    local2()    
    print(x)
local1()
print(x)
```

```python
x = 'fulano'

def escopo1():
    x = 'sicrano'
    print('primeira chamada {}'.format(x))
    def escopo2():
        global x
        print('segunda chamada {}'.format(x))
        x = 'beltrano'
        print('terceira chamada {}'.format(x))
        #print(x)
        return x
    print('quarta chamada {}'.format(x))
    x = escopo2()
    print('quinta chamada {}'.format(x))
    
    
print('sexta chamada {}'.format(x))
x = escopo1()
print('setima chamada {}'.format(x))
```

Recordando: métodos de tipos mutáveis e métodos de tipos imutáveis

```python
lista = [1,2,3]

def acrescenta(uma_lista):
    uma_lista.append(4)
    return uma_lista

lista = acrescenta(lista)
print(lista)
```

```python
lista = [1,2,3]

def acrescenta(uma_lista):
    uma_lista.append(4)

acrescenta(lista)
print(lista)
```

```python
lista = [1,2,3]

def acrescenta():
    lista.append(4)

acrescenta()
print(lista)
```

```python
def scope_test():
    def do_local():
        spam = "local spam"
        
    def do_nonlocal():
        nonlocal spam
        spam = "nonlocal spam"
        
    def do_global():
        global spam
        spam = "global spam"
        
    spam = "test spam"
    do_local()

    print("After local assignment:", spam)
    do_nonlocal()
    print("After nonlocal assignment:", spam)
    do_global()
    print("After global assignment:", spam)

scope_test()
print("In global scope:", spam)
```
